```python
#library import
import numpy as np
import pandas as pd
from sklearn import tree

#read data
vgdata = pd.read_csv('vgsales.csv', delimiter=',', header=0)
```

```python
vgdata.head()
```

```python
vgdata["Platform"] = pd.factorize(vgdata.Platform)[0]
```

```python
vgdata.head()
```

```python
print(vgdata)
```

```python
vgdata = vgdata.drop(labels="Name", axis=1)
vgdata = vgdata.drop(labels="Rank", axis=1)
vgdata = vgdata.drop(labels="Genre", axis=1)
vgdata = vgdata.drop(labels="Publisher", axis=1)
print(vgdata)
```

```python
vgdata = vgdata.to_numpy()
```

```python
dataTraining = np.concatenate((vgdata[0:12500, :], vgdata[13500:15000]),axis=0)
dataTesting = np.concatenate((vgdata[12500:13500, :], vgdata[15000:16598]), axis=0)
dataTraining = dataTraining.astype('int')
dataTesting = dataTesting.astype('int')
```

```python
inputTraining = dataTraining[:, 0:7]
inputTesting = dataTesting[:, 0:7]
labelTraining = dataTraining[:, 6]
labelTesting = dataTesting[:, 6]
#print(dataTraining)
```

```python
model = tree.DecisionTreeClassifier()
model = model.fit(inputTraining, labelTraining)
```

```python
hasilPrediksi = model.predict(inputTesting)
print("Label sebenarnya ", labelTesting)
print("Hasil Prediksi : ", hasilPrediksi)

prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("Prediksi Benar : ", prediksiBenar, "data.")
print("Prediksi Salah : ", prediksiSalah, "data.")
print("Akurasi : ", prediksiBenar/(prediksiBenar+prediksiSalah)*100, "%")
```
